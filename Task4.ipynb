Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False

    return True


# Main Program
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))

        if validate_recharge(amount):
            print("Recharge Successful!")
            break
        else:
            print("Please try again.\n")

    except ValueError:
        print("Please enter a valid number.\n")
```

Problem 2: Inventory Reorder Alert System


```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} → Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} → Stock OK (Stock: {stock})")


# Sample Inventory Data
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Wheat": 15,
    "Salt": 30,
    "Pepper": 25
}

# Function Call
check_inventory(inventory)
```


 Problem 3: Student Result Processing System


```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} → Pass"
    else:
        return f"Average: {average:.2f} → Fail"


# Example Marks
student_marks = [60, 45, 70, 55]

# Function Call
result = process_result(student_marks)
print(result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


# Main Program with Retry Option
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator")
            break

    except ValueError:
        print("Please enter valid input!\n")
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance):
    present_days = 0

    # Count present days using loop
    for day in attendance:
        if day == "P":
            present_days += 1

    total_days = len(attendance)
    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% → Eligible"
    else:
        return f"Attendance: {percentage:.2f}% → Not Eligible"


# Example Attendance Record
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Function Call
result = check_eligibility(attendance_list)
print(result)
```

Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    has_digit = False
    has_special = False

    # Check each character using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    # Conditions
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    elif not has_digit:
        return "Weak Password (Must contain at least one digit)"
    elif not has_special:
        return "Weak Password (Must contain at least one special character @#$)"
    else:
        return "Strong Password"


# Example
pwd = input("Enter your password: ")
result = check_password_strength(pwd)
print(result)
```
